##Ejercicio 1.

Diseñe un programa que le permita a un usuario:     

*   Crear una lista de palabras tan larga como quiera.
*   El programa debe mostrar las palabras de la lista que tienen longitudes $\geq 3$, es decir, con tres o más letras.



##Ejercicio 2.

Diseñe un programa que le permita a un usario crear:

* Dos listas de palabras **A** y **B** de igual longitud, pero con las palabras que el usuario desee. 

* Un diccionario **D** cuyas claves correspondan a las palabras en la lista **A** y cuyos valores correspondan a los de la lista **B**. De tal modo que, la primera palabra en **A** tenga como valor asociado la primera palabra en **B**, y así sucesivamente. Por ejemplo, si la lista **A** es:

**A** =[Nombre, Apellido, Identificación, Edad, Dirección]

y las lista **B** es:

**B** = [Luis, Garcia, 1213234, 23, k56#10-20]

El diccionario debe contener:

**D** = { 'Nombre':'Luis', 'Apellido':'Garcia', 'Identificación':'1213234', 'Edad':'23', 'Dirección':'k56#10-20'}

**Nota:** El programa debe funcionar para cualquier colección de palabras que brinde el usario.

# Ejercicio 7

Diseñe un programa que haga uso de una función que permita construir un diccionario llamado “Biblioteca” en donde se guarde la información de los libros de una biblioteca con las siguientes características:

a) El número de ítems en el diccionario, lo da el usuario.

b) La clave de cada ítem, corresponde al string “Libro_k”, donde k es un número dado por el usuario.

c) El valor de cada ítem, es un diccionario con las siguientes claves: Título, Autor, Año y Editorial.

Por ejemplo: Supongamos que los libros que se guardan son:

    Libro 1: Autor = Gabriel García Márquez, Título = Cien años de soledad, Año = 1968, Editorial = Oveja Negra.

    Libro 2: Autor = Julio Cortázar, Título=Rayuela, Año=1963, Editorial = Sudamericana.

    Libro 3: Autor = Jorge Luis Borges, Título=El Aleph, Año=1945, Editorial = Emece.

Entonces, el diccionario Biblioteca debe contener:

Biblioteca = { Libro 1: {Autor : Gabriel García Márquez, Título : Cien años de soledad, Año : 1968, Editorial = Oveja Negra}, Libro 2: {Autor : Julio Cortázar, Título : Rayuela, Año=1963, Editorial = Sudamericana} , Libro 3: {Autor : Jorge Luis Borges, Título : El Aleph, Año=1945, Editorial = Emece}}.
