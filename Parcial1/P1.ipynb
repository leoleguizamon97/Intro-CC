# Ejercicio 1. (20 Ptos.)

Diseñe un programa que reciba una lista de conjuntos de números dados por un usuario y que calcule:  la intersección y la unión  de los conjuntos presentes en la lista. El número de conjuntos en la lista lo da el usuario.

Por ejemplo, si la lista de conjuntos ingresada por el usuario es:

$L = [\{1,2,5\}, \{7,8,1,13,2\},\{3,5,9,0,1,2\}, \{1,3,14,2,15\} ]$

El programa debe reportar:

La intersección es: $\{1,2\}$ .

La unión es $\{0,1,2,3,5,7,8,9,13,14,15\}$ .



```python
#ciclo recibir conjuntos
numConjuntos = int(input("Ingrese la cantidad de conjuntos: "))

conjuntos = []

for i in range(numConjuntos):
    #Ciclo conjunto 1
    print(f"Ingrese los elementos del conjunto {i+1}")
    conjunto1 = set()
    contador = 0
    while(contador != -1):
        contador += 1
        conjunto1.add(int(input(f"Ingrese el elemento NUMERICO #{contador}: ")))
        #Valida intencion de usuario
        continuar = input("¿Desea ingresar otro elemento? (n): ")
        if continuar == "n" or continuar == "N":
            conjuntos.append(conjunto1)
            contador = -1

#hacemos el ciclo para definir las uniones

union = conjuntos[0]
interseccion = conjuntos[0]

print(union)

for a in conjuntos:
    union = union | a
    interseccion = interseccion & a


#Mostrar resultados
print(f"Union {union}, Interseccion {interseccion}")
```

# Ejercicio 5. (20 Ptos)

Diseñe un programa  que permita construir un diccionario llamado “Restaurante” en donde se guarde la información de los menús del restaurante con las siguientes características:

**(A)** El número de ítems en el diccionario, lo da el usuario.

**(B)** La clave de cada ítem, corresponde al string “Menu_k”, donde k es un número dado por el usuario.

**(C)** El valor de cada ítem, es un diccionario con las siguientes claves: Entrada, PlatoPrincipal, Bebida y Postre.

Por ejemplo: Supongamos que los menus son:

    Menu 1: {Entrada : Frutas, PlatoPrincipal : Pasta Napolitana, Bebida : Vino Tinto, Postre = Tres Leches}.

    Menu 2: {Entrada : Brochetas de Pollo, PlatoPrincipal :Costillas de Cerdo, Bebida = Cerveza, Postre = Natas}.

    Menu 3: {Entrada : Coctel de Camarones, PlatoPrincipal:Asado de Res, Bebida :Vino Blanco , Postre = Helado de Vainilla}.

Entonces, el diccionario Restaurante debe contener:

**Restaurante** = { **Menu 1**: {Entrada : Frutas, PlatoPrincipal : Pasta Napolitana, Bebida : Vino Tinto, Postre = Tres Leches.}}, **Menu 2**: {Entrada : Brochetas de Pollo, PlatoPrincipal :Costillas de Cerdo, Bebida = Cervez, Postre = Natas} , **Menu 3**: {Entrada : Coctel de Camarones, PlatoPrincipal:Asado de Res, Bebida :Vino Blanco , Postre = Helado de Vainilla}}.


```python
#menu Entrada, PlatoPrincipal, Bebida y Postre.
cantidadMenus = int(input("Ingrese la cantidad de menus que desea: "))
menuBase = {"Entrada":"","Plato principal":"","Bebida":"","Postre":"",}
restaurante = {}
for i in range(cantidadMenus):
    menu = {}
    print(f"Ahora ingresará los elementos del menu {i+1}")
    for clave in menuBase:
        elemento = input(f"Ingrese el/la {clave} del menu {i+1}: ")
        menu[clave] = elemento
    nombreMenu = "Menu " + str(i+1)
    restaurante[nombreMenu] = menu

print(f"Acá todos los menus: \n {restaurante}")
```
