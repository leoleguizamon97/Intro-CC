**Ejercicio 1.** Considere una cadena (camino) poligonal (https://es.wikipedia.org/wiki/Cadena_poligonal). Diseñe un código que determine un par de vértices contiguos o sucesivos en el camino poligonal que estén más separados. Tenga en cuenta los siguientes requerimientos para el código:

* Las coordenadas de los puntos del camino las da el usuario, en el orden en que se recorre el camino. Las coordenadas deben guardarse en una lista.
* El programa debe tener una función que calcula la distancia entre dos puntos del plano cartesiano y esta función debe ser de tipo lambda.
* El programa debe tener una función que determina un par de vértices contiguos más separados de camino poligonal.
* El programa debe informar cual fue el par de vertices escogido y la distancia entre ellos.

**Ejercicio 5.** Diseñe un código que lea los coeficientes de un par de rectas de la forma $ax + by = c$ desde un archivo de texto llamado Datos.txt. y calcule el punto de corte de las rectas, en el caso en que éste exista, o indique si las rectas son paralelas y no se cortan. En este caso debe leer 6 datos correspondientes a los valores de a, b y c para cada una de las rectas. Los datos deben disponerse por filas, por ejemplo, si las rectas son: $4x + 3y = 1$ y $-3x + 5y = 3$.  En el archivo de texto, deben aparecer los datos en el siguiente orden:

 4 3 1

-3 5 3

El reporte de resultados debe suministrarse en otro archivo denominado Resultados.txt. El reporte debe mostrar los mensajes:

* Las rectas $ax+by = C$ y $dx+ey = f$ se cortan en el punto $(x_0,y_0)$.
* Las rectas $ax+by = C$ y $dx+ey = f$ no se cortan y por lo tanto son paralelas.
